```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from mpl_toolkits.mplot3d import Axes3D
from pylab import figure
```

```python
# задание 1

x = (10, 10, 10)
y = (0, 0, -10)

z = []
for i in range(len(x)):
    z.append(x[i] + y[i])
print('Вектор x =', x)
print('Вектор y =', y)
print('Вектор x + y =', tuple(z))

sum_z = 0
for i in range(len(z)):
    sum_z += z[i]**2
len_z = sqrt(sum_z)
print('Длина вектора x + y =', round(len_z, 4))
```

```python
# задание 3: окружность

x = np.linspace(-10, 10, 1001)
R = 10
a = 0
b = 10

y1 = [sqrt(R**2 - (x[i] - a)**2) + b for i in range(1001)]
y2 = [-sqrt(R**2 - (x[i] - a)**2) + b for i in range(1001)]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# задание 3: эллипс

x = np.linspace(-10, 10, 1001)
a = 10
b = 5

y1 = [sqrt(1 - x[i]**2 / a**2) * b for i in range(1001)]
y2 = [-sqrt(1 - x[i]**2 / a**2) * b for i in range(1001)]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-10, 10)
plt.grid()
```

```python
# задание 3: гипербола

x1 = np.linspace(-10, -3, 1001)
x2 = np.linspace(3, 10, 1001)

a = 3
b = 10

y1 = [sqrt(x1[i]**2 / a**2 - 1) * b for i in range(1001)]
y2 = [-sqrt(x1[i]**2 / a**2 - 1) * b for i in range(1001)]
y3 = [sqrt(x2[i]**2 / a**2 - 1) * b for i in range(1001)]
y4 = [-sqrt(x2[i]**2 / a**2 - 1) * b for i in range(1001)]

plt.plot(x1, y1)
plt.plot(x1, y2)
plt.plot(x2, y3)
plt.plot(x2, y4)

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# задание 5.1

ax = Axes3D(figure())

x = np.linspace(-10, 10, 11)
y = np.linspace(-10, 10, 11)
x, y = np.meshgrid(x, y)

z1 = 3 * x + 4 * y - 50
z2 = 3 * x + 4 * y + 50

ax.plot_surface(x, y, z1)
ax.plot_surface(x, y, z2)
plt.show()
```

```python
# задание 5.2

ax = Axes3D(figure())

x = np.linspace(-10, 10, 31)
y = np.linspace(-10, 10, 31)
x, y = np.meshgrid(x, y)

z1 = 10 * x**2 + 5 * y**2 + 100
z2 = 5 * x**2 - 10 * y**2 - 100

ax.plot_surface(x, y, z1)
ax.plot_surface(x, y, z2)
plt.show()
```
