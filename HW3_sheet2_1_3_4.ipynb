```python
# задание 4.1: система уравнений

x = np.linspace(-5, 5, 101)

np.warnings.filterwarnings('ignore')

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 2))
print(f'x1 = {x1}\ny1 = {y1}\n')

x2, y2 = fsolve(equations, (0, -1))
print(f'x2 = {x2}\ny2 = {y2}\n')

x3, y3 = fsolve(equations, (3, 6))
print(f'x3 = {x3}\ny3 = {y3}\n')

x4, y4 = fsolve(equations, (4, 15))
print(f'x4 = {x4}\ny4 = {y4}\n')


plt.plot(x, 1 - ((1 - np.exp(x)) / x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```

```python
# задание 4.1: система неравенств

x = np.linspace(-5, 5, 101)


def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 2))
x2, y2 = fsolve(equations, (0, -1))
x3, y3 = fsolve(equations, (3, 6))
x4, y4 = fsolve(equations, (4, 15))

x = np.linspace(-5, x1, 101)
plt.plot(x, x**2 - 1, color='orange')

x = np.linspace(x2, x3, 101)
plt.plot(x, x**2 - 1, color='orange')

x = np.linspace(x4, 5, 101)
plt.plot(x, x**2 - 1, color='orange')

print (f'x ∈ (-∞; {x1}), ({x2}; {x3}), ({x4}; +∞)')

plt.xlabel('x')
plt.ylabel('y')
plt.grid()
```
