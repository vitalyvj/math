```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
from scipy.optimize import fsolve
```

```python
# задание 1

x = np.linspace(-10, 10, 1001)

k1 = 1
a1 = 2
b1 = 3

k2 = 3
a2 = 2
b2 = 1

k3 = 4
a3 = 5
b3 = -2

plt.plot(x, k1 * np.cos(x - a1) + b1)
plt.plot(x, k2 * np.cos(x - a2) + b2)
plt.plot(x, k3 * np.cos(x - a3) + b3)

plt.xlabel('x')
plt.ylabel('y (x) = k∙cos(x – a) + b')
plt.grid()
```

```python
# задание 4.1: система уравнений

x = np.linspace(-5, 5, 101)

np.warnings.filterwarnings('ignore')

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 2))
print ('x1 =', x1, 'y1 =', y1)

x2, y2 = fsolve(equations, (3, 6))
print ('x2 =', x2, 'y2 =', y2)

x3, y3 = fsolve(equations, (4, 15))
print ('x3 =', x3, 'y3 =', y3)


plt.plot(x, 1 - ((1 - np.exp(x)) / x))
plt.plot(x, x**2 - 1)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5, 5)
plt.ylim(-1, 30) 
plt.grid()
```

```python
# задание 4.1: система неравенств

x = np.linspace(-5, 5, 101)


def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 2))

x2, y2 = fsolve(equations, (3, 6))

x = np.linspace(-5, x1, 101)
plt.plot(x, x**2 - 1, color='orange')

x = np.linspace(x2, x3, 101)
plt.plot(x, x**2 - 1, color='orange')

print (f'x ∈ (-∞; {x1}), ({x2}; {x3})')


plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5, 5) 
plt.ylim(-1, 30) 
plt.grid()    
```
