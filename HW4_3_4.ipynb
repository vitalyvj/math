```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
# задание 3

a = 10
b = 15
alpha = np.pi / 6
x0 = 20
y0 = 11

x = np.linspace(x0, x0 + b * np.cos(alpha), 101)
y = np.tan(alpha) * (x - x0) + y0
plt.plot(x, y)
plt.grid()

for i in range(4):
    plt.plot([0, 100],[a * i, a * i], linewidth=0.75, color ='red', linestyle ="--")
    if y0 >= a * i:
        n = i + 1

if y0 < n * a - b * np.sin(alpha):
    print('Игла линию не пересекает')
else:
    print('Игла пересекает линию')
```
